```python
import scipy
```

```python
from scipy import fftpack
```

```python
dw = 1/300.0
time = scipy.arange(0,1024)*dw
freq = scipy.linspace(-150,150,1024)
```

```python
w1 =  2.0*pi*100.0
w2 =  2.0*pi*140.0
ccc = scipy.cos( time*w1)*scipy.exp(-time*2.0)
ccc += scipy.cos( time*w2)*scipy.exp(-time*2.0)
print pi
```

```python
sss = scipy.sin( time*w1)*scipy.exp(-time*2.0)
sss += scipy.sin( time*w2)*scipy.exp(-time*2.0)
```

```python
plot( ccc );
```

```python
sss_i = fftpack.hilbert(ccc)
```

```python
#plot(ccc[:20], 'r');
plot(sss[:20]);
plot(-1*sss_i[:20]);
```

```python
sss_r = fftpack.fft(ccc)
sss_ir = fftpack.fft((-1)*sss_i)
```

```python
plot( freq, fftpack.fftshift(sss_r.real), 'r' );
#plot( fftpack.fftshift(sss_ir.real) );
```
